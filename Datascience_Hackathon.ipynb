### Checking and creating a synthetic target variable if missing

```python
if 'exam score' not in train_data.columns:
    train_data['exam score'] = (train_data['practical score'] + train_data['viva score']) // 2
```

```python
# Defining features and target variable
features = ['test preparation ', 'gender', 'parental level of education', 'lunch', 'Section', 'practical score', 'viva score']
X_train = train_data[features]
y_train = train_data['exam score']
```

```python
X_test = test_data[features]
```

```python
# Converting categorical columns to numeric using Label Encoding
label_encoder = LabelEncoder()
for column in ['gender', 'parental level of education', 'lunch', 'Section', 'test preparation ']:
    if column in X_train.columns:
        X_train[column] = label_encoder.fit_transform(X_train[column])
    if column in X_test.columns:
        X_test[column] = label_encoder.transform(X_test[column])
```

### Random Forest Classifier

```python
# Initializing and training the model
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
# Predict on the test data
test_predictions = clf.predict(X_test)
```

### Preparing the submission

```python
submission = pd.DataFrame({
    'Roll no': test_data['Roll no'],
    'exam score': test_predictions})
```

### Saving the submission file

```python
submission.to_csv('Datascience_Hackathon1.csv', index=False)
print('Submission file created: Datascience_Hackathon.csv')
```
